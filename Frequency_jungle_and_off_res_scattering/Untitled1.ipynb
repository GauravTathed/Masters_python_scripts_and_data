```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

def plot_d52_calibration(dates):
    plt.figure(figsize=(18, 6))
    
    for date in dates:
        # Define the file pattern based on the given date
        file_pattern =  f"Z:\Lab Data\D52_Calibration_Ba137\Complete_Calibration_Raw_data\D52_calibration_raw_data_*_{date}_*.txt"
        files = sorted(glob.glob(file_pattern))
        
        if not files:
            print(f"No files found for date {date}.")
            continue
        
        for file in files:
            with open(file, 'r') as f:
                lines = f.readlines()
            
            # Ignore the last line
            data = np.loadtxt(lines[:-1], delimiter=",", skiprows=0)
            
            x = data[:, 0]
            y = data[:, 1]
            plt.plot(x, y, marker=".", linestyle="-", label=f"{os.path.basename(file)} ({date})")
    
    plt.xlabel("Frequency (MHz)")
    plt.ylabel("Signal Intensity")
    plt.title("D52 Calibration Data for Multiple Dates")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Example usage:
dates = ["20250303", "20250304"]  # Change to desired dates
plot_d52_calibration(dates)
```
