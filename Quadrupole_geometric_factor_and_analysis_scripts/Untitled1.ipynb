```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
E0 = 1               # Amplitude of the electric field
chi = 1      # Ellipticity angle (adjust between 0 and pi/4 for general cases)
phi = np.pi / 2      # Phase difference
alpha = np.pi/4   # Orientation angle of the ellipse
t = np.linspace(0, 2 * np.pi, 100)  # Time parameter

# Electric field components
Ex = E0 * (np.cos(chi) * np.cos(alpha) * np.cos(t) - np.sin(chi) * np.sin(alpha) * np.cos(t + phi))
Ey = E0 * (np.cos(chi) * np.sin(alpha) * np.cos(t) + np.sin(chi) * np.cos(alpha) * np.cos(t + phi))

# Plotting the polarization ellipse
plt.figure(figsize=(8, 8))
plt.plot(Ex, Ey, label="Polarization Ellipse")
plt.axhline(0, color="gray", lw=0.5)
plt.axvline(0, color="gray", lw=0.5)
plt.xlabel("E_x")
plt.ylabel("E_y")
plt.title("Polarization Ellipse")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()
```
