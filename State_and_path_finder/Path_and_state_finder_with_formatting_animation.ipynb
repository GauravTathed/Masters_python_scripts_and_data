```python
from collections import deque
import numpy as np

def find_least_weight_paths(start_node, end_node, matrix = pi_times, weight_matrix = pi_times, min_weight=0):
    G = plot_coupling_graph_from_matrix(matrix, weight_matrix, show_graph=False)
    queue = deque([[start_node]])
    least_weight_paths = []
    best_weight = None  # Delay initializing the best weight until we find a valid path

    while queue:
        path = queue.popleft()
        node = path[-1]

        # Compute the current path's total weight
        current_weights = [G[path[i]][path[i + 1]]['weight'] for i in range(len(path) - 1)]
        current_total_weight = np.sum(current_weights)

        # If best_weight is already set, prune paths exceeding it
        if best_weight is not None and current_total_weight > best_weight:
            continue

        if node == end_node:
            if best_weight is None or current_total_weight < best_weight:
                # Set or update the best weight when a valid path is found
                best_weight = current_total_weight
                least_weight_paths = [(path, current_total_weight, current_weights)]
            elif current_total_weight == best_weight:
                # Add paths with the same minimum weight
                least_weight_paths.append((path, current_total_weight, current_weights))
        else:
            for neighbor in G.neighbors(node):
                if neighbor not in path:  # Avoid cycles
                    edge_weight = G[node][neighbor]['weight']
                    if edge_weight >= min_weight:
                        # Compute estimated total weight
                        estimated_total_weight = current_total_weight + edge_weight

                        if best_weight is None or estimated_total_weight <= best_weight:
                            new_path = list(path)
                            new_path.append(neighbor)
                            queue.append(new_path)

    return least_weight_paths
```

```python
def plot_coupling_graph_from_matrix(matrix, transition_strengths=None, pos=None, show_graph=True):
    import networkx as nx
    import matplotlib.pyplot as plt
    import matplotlib.colors as mcolors
    from matplotlib.patches import Rectangle
    import numpy as np  # Ensure numpy is imported

    rows, cols = matrix.shape

    # Create a graph
    G = nx.Graph()

    # Combine row_labels and col_labels into a single labels list
    labels = []

    Fs = [1, 2, 3, 4]
    for i in Fs:
        for j in range(2 * i + 1):
            mF = i - j
            labels.append([i, mF])

    col_labels = [-2, -1, 0, 1, 2]
    for c in col_labels:
        labels.append(c)

    # Add edges based on non-zero elements in the matrix and assign weights
    for i in range(rows):
        for j in range(cols):
            if matrix[i, j] != 0 and i!=j:
                G.add_edge(f"{labels[i]}", f"{labels[j]}", weight=transition_strengths[i, j])

    # Generate positions for the nodes
    if pos is None:
        pos = {}
        zeeman_coefficient = 0.1  # Adjust this value to control the amount of splitting
        for idx, label in enumerate(labels):
            if isinstance(label, list):
                F, mF = label
                x = mF
                # Base y-position based on F level
                y_base = abs(F - 4)
                # Displacement proportional to mF to simulate Zeeman splitting
                y_displacement = zeeman_coefficient * mF
                y = y_base + y_displacement
                pos[f"{label}"] = (x, y)
            else:
                # label is an integer (col_labels)
                x = label
                y = -2  # Keep the same y-position for col_labels
                pos[f"{label}"] = (x, y)

    # Add nodes to the graph with positions
    for label in labels:
        G.add_node(f"{label}", pos=pos[f'{label}'])

    if show_graph:
        plt.figure(figsize=(12, 9))

        edges = list(G.edges())
        weights = np.array([G[u][v]['weight'] for u, v in edges])

        # Normalize the weights to range between 0 and 1
        norm = plt.Normalize(vmin=np.min(weights), vmax=np.max(weights))

        # Choose the 'viridis' colormap
        cmap = plt.cm.viridis

        # Map the weights to colors
        edge_colors = cmap(norm(weights))

        nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=edge_colors, width=2,alpha = (1-norm(weights)))

        # Draw nodes with rectangular patches
        xh = 0.5
        yh = 0.25
        ax = plt.gca()
        for node, (x, y) in pos.items():
            ax.add_patch(Rectangle((x - xh / 2, y - yh / 2), xh, yh,
                                   fill=True, color='black', ec='black', lw=1))

        nx.draw_networkx_labels(G, pos, font_size=10, font_color='white')

        # Adjust the axes limits to accommodate the displaced nodes
        plt.xlim(-4.5, 4.5)
        plt.ylim(-3, 4)  # Adjusted ylim to fit the displaced nodes

        # Show the plot
        plt.title('Coupling Graph from Matrix with Zeeman Splitting')
        plt.show()

    return G
```

```python
import time
import random
from itertools import combinations

def find_original_path(list_of_paths, start_node, end_node):

    for path in list_of_paths:
        if start_node in path and end_node in path:
            start_idx = path.index(start_node)
            end_idx = path.index(end_node)
            if start_idx < end_idx:  # Ensure the path flows from start to end
                return path[start_idx:end_idx + 1]
    return None

def all_nodes_involved(nodes, list_of_paths=list_of_paths_og):

    if not nodes:
        return []  # Return empty list if no nodes provided

    pairs = list(combinations(nodes, 2))
    unique_nodes = set()
    # print(pairs)
    for start_node, end_node in pairs:
        path = find_original_path(list_of_paths, start_node, end_node)
        # print(find_original_path(list_of_paths, start_node, end_node), start_node, end_node)
        # print(shfkjh)
        if path: 
            unique_nodes.update(path)

    return list(unique_nodes)


def compute_cost_bound(full_nodes_set):

    dpb = len(full_nodes_set)
    if dpb == 0:  # Prevent division by zero
        return float('inf')

    subG_kB = GLOBAL_G_kB.subgraph(full_nodes_set)
    subG_L_G = GLOBAL_G_L_G.subgraph(full_nodes_set)
    subG_L_L = GLOBAL_G_L_L.subgraph(full_nodes_set)
    subG_pi = GLOBAL_G_pi_times.subgraph(full_nodes_set)

    sum_kB = sum(data.get('weight', 1) for u, v, data in subG_kB.edges(data=True))
    sum_L_G = sum(data.get('weight', 1) for u, v, data in subG_L_G.edges(data=True))
    sum_L_L = sum(data.get('weight', 1) for u, v, data in subG_L_L.edges(data=True))
    sum_pi = sum(data.get('weight', 1) for u, v, data in subG_pi.edges(data=True))

    cost = ((sum_L_L * sum_pi + sum_L_G * (sum_pi ** 2) + sum_kB * (sum_pi ** 2)) / (dpb**2))
    return cost


def branch_and_bound(d, min_cost=float('inf'), best_set=None):

    remaining_nodes = list(G_pi_times.nodes())
    random.shuffle(remaining_nodes)

    initial_pairs = list(combinations(remaining_nodes, 1))[::-1]
    
    for pair in initial_pairs:
        current_set = list(pair)
        # print(current_set)
        remaining_nodes_after_pair = [node for node in remaining_nodes if node not in pair]

        min_cost, best_set = _branch_and_bound_recursive(
            d, current_set, remaining_nodes_after_pair, min_cost, best_set
        )
        # print(min_cost, best_set)
    return min_cost, best_set


def _branch_and_bound_recursive(d, current_set, remaining_nodes, min_cost, best_set):

    if len(current_set) == d:
        full_nodes_set = all_nodes_involved(current_set)
        cost = compute_cost_bound(full_nodes_set)
        if cost < min_cost:
            min_cost = cost
            best_set = list(current_set)
        return min_cost, best_set

    for idx, node in enumerate(remaining_nodes):
        # print(node, min_cost)
        new_set = current_set + [node]
        new_remaining_nodes = remaining_nodes[idx + 1 :]

        full_nodes_set = all_nodes_involved(new_set)
        estimated_cost = compute_cost_bound(full_nodes_set)

        if estimated_cost >= min_cost:
            continue

        min_cost, best_set = _branch_and_bound_recursive(
            d, new_set, new_remaining_nodes, min_cost, best_set
        )
        # print(best_set)
    return min_cost, best_set


d =  4 # Number of nodes to select
start_time = time.time()  # Start timer
min_cost, best_nodes = branch_and_bound(d)
end_time = time.time()  # End timer

print(f"Best combination of nodes: {best_nodes}")
print(f"Minimum cost: {min_cost}")
print(f"Time taken: {end_time - start_time:.2f} seconds")
```

```python
import time
import random
import math
from itertools import combinations

# =============================================================================
# Utility Functions (assumes list_of_paths_og and global graphs are defined)
# =============================================================================

B = 4.7e-5
kB = (np.abs(Sen_matrix)*B)**2
L_G = (1/2054.4241218909024)**2
L_L = 1/1916.8500604713997

L_LN_matrix = np.full((29, 29), L_L)
L_GN_matrix = np.full((29, 29), L_G)

L_LN_matrix[0:24, 0:24] = 0  
L_LN_matrix[-5:, -5:] = 0 

L_GN_matrix[0:24, 0:24] = 0 
L_GN_matrix[-5:, -5:] = 0  
G_AC_line = plot_coupling_graph_from_matrix(np.abs(Sen_matrix), np.abs(Sen_matrix),show_graph=False)
G_kB = plot_coupling_graph_from_matrix(kB, kB,show_graph=False)
G_L_G = plot_coupling_graph_from_matrix (L_GN_matrix,L_GN_matrix,show_graph=False)
G_L_L = plot_coupling_graph_from_matrix (L_LN_matrix,L_LN_matrix,show_graph=False)
G_pi_times = plot_coupling_graph_from_matrix(pi_times_array, pi_times_array,show_graph=False)
G_buses = plot_coupling_graph_from_matrix(sen_sq_matrix_29x29*(B**2), sen_sq_matrix_29x29*(B**2),show_graph=False)
G_infidelity = plot_coupling_graph_from_matrix(infidelity_matrix_29x29, infidelity_matrix_29x29,show_graph=False)

# Define global variables for graphs
GLOBAL_AC_Line = G_AC_line
GLOBAL_G_kB = G_kB
GLOBAL_G_L_G = G_L_G
GLOBAL_G_L_L = G_L_L
GLOBAL_G_pi_times = G_pi_times

def find_original_path(list_of_paths, start_node, end_node):
    """
    Given a list of paths, returns the segment from start_node to end_node
    if such a segment exists in any of the paths.
    """
    for path in list_of_paths:
        if start_node in path and end_node in path:
            start_idx = path.index(start_node)
            end_idx = path.index(end_node)
            if start_idx < end_idx:  # Ensure the path flows from start to end
                return path[start_idx:end_idx + 1]
    return None

def all_nodes_involved(nodes, list_of_paths=list_of_paths_og):
    """
    Given a list of nodes, for every pair of nodes, if there is a connecting path,
    include all nodes from that path. Returns the union (unique set) of nodes.
    """
    if not nodes:
        return []
    pairs = list(combinations(nodes, 2))
    unique_nodes = set()
    for start_node, end_node in pairs:
        path = find_original_path(list_of_paths, start_node, end_node)
        if path:
            unique_nodes.update(path)
    return list(unique_nodes)

def compute_cost_bound(full_nodes_set):
    """
    Computes the cost of the given set of nodes using the subgraphs of the global graphs.
    """
    dpb = len(full_nodes_set)
    if dpb == 0:  # Prevent division by zero
        return float('inf')
    
    # Obtain subgraphs for the given set of nodes.
    subG_kB  = GLOBAL_G_kB.subgraph(full_nodes_set)
    subG_L_G = GLOBAL_G_L_G.subgraph(full_nodes_set)
    subG_L_L = GLOBAL_G_L_L.subgraph(full_nodes_set)
    subG_pi  = GLOBAL_G_pi_times.subgraph(full_nodes_set)
    sub_ac = GLOBAL_AC_Line.subgraph(full_nodes_set)
    # Sum weights from each subgraph (defaulting to 1 if the edge attribute 'weight' isnt set)
    sum_kB  = sum(data.get('weight', 1) for u, v, data in subG_kB.edges(data=True))
    sum_L_G = sum(data.get('weight', 1) for u, v, data in subG_L_G.edges(data=True))
    sum_L_L = sum(data.get('weight', 1) for u, v, data in subG_L_L.edges(data=True))
    sum_pi  = sum(data.get('weight', 1) for u, v, data in subG_pi.edges(data=True))
    sum_ac = sum(data.get('weight', 1) for u, v, data in sub_ac.edges(data=True))
    cost = ((sum_L_L * sum_pi + sum_L_G * (sum_pi ** 2) + sum_kB * (sum_pi ** 2)) + sum_ac * (sum_pi**2)) 
    return cost

# =============================================================================
# Simulated Annealing Function
# =============================================================================

def simulated_annealing(d, initial_temp=100.0, cooling_rate=0.995, max_iter=10000):
    """
    Uses simulated annealing to select d nodes (from G_pi_times) that minimize the cost.
    
    Parameters:
      - d: Desired number of nodes in the final solution.
      - initial_temp: Starting temperature.
      - cooling_rate: Rate at which the temperature decreases.
      - max_iter: Maximum number of iterations.
    
    Returns:
      - best_solution: The best set of nodes found.
      - best_cost: The associated cost of best_solution.
    """
    # Get the list of all available nodes from G_pi_times.
    nodes_list = list(G_pi_times.nodes())
    if d > len(nodes_list):
        raise ValueError("d cannot be greater than the total number of nodes in G_pi_times.")

    # Start with a random initial solution (select d random nodes).
    current_solution = random.sample(nodes_list, d)
    current_full_set = all_nodes_involved(current_solution)
    current_cost = compute_cost_bound(current_full_set)
    
    best_solution = list(current_solution)
    best_cost = current_cost
    
    temperature = initial_temp

    for iteration in range(max_iter):
        # Generate a neighbor solution by swapping one node.
        neighbor_solution = list(current_solution)
        idx_to_replace = random.randrange(d)
        # Pick a candidate node that is not already in the current solution.
        candidate_nodes = [node for node in nodes_list if node not in neighbor_solution]
        if not candidate_nodes:
            break  # Should not happen if d < total nodes.
        new_node = random.choice(candidate_nodes)
        neighbor_solution[idx_to_replace] = new_node

        # Evaluate the neighbor solution.
        neighbor_full_set = all_nodes_involved(neighbor_solution)
        neighbor_cost = compute_cost_bound(neighbor_full_set)
        
        # Determine whether to accept the neighbor solution.
        delta = neighbor_cost - current_cost
        if delta < 0:
            # Better solution found; accept it.
            current_solution = neighbor_solution
            current_cost = neighbor_cost
            if neighbor_cost < best_cost:
                best_solution = list(neighbor_solution)
                best_cost = neighbor_cost
        else:
            # Accept with a probability based on the temperature.
            if random.random() < math.exp(-delta / temperature):
                current_solution = neighbor_solution
                current_cost = neighbor_cost
        
        # Update the temperature (cooling).
        temperature *= cooling_rate
        if temperature < 1e-8:
            break  # Temperature is too low to continue meaningful exploration.

    return best_solution, best_cost

# =============================================================================
# Iterative Simulated Annealing Runs
# =============================================================================

def iterative_simulated_annealing(d, num_runs=100, **sa_params):
    """
    Runs the simulated annealing procedure num_runs times and returns the best
    solution (i.e., the one with the minimum cost) among all runs.
    
    Parameters:
      - d: Desired number of nodes in the final solution.
      - num_runs: Number of times to run simulated annealing.
      - sa_params: Additional parameters for simulated annealing (initial_temp, cooling_rate, max_iter).
      
    Returns:
      - global_best_solution: The best set of nodes found over all runs.
      - global_best_cost: The associated cost of the best solution.
    """
    global_best_cost = float('inf')
    global_best_solution = None
    
    for run in range(num_runs):
        solution, cost = simulated_annealing(d, **sa_params)
        if cost < global_best_cost:
            global_best_cost = cost
            global_best_solution = solution
        # Optionally, print progress information:
        print(f"Run {run+1}/{num_runs}: Cost = {cost:.4f}, Best so far = {global_best_cost:.4f}")
    
    return global_best_solution, global_best_cost

# =============================================================================
# Example Usage
# =============================================================================

d = 10  # Desired number of nodes
num_runs = 5  # Number of simulated annealing runs

start_time = time.time()

best_nodes, min_cost = iterative_simulated_annealing(
    d,
    num_runs=num_runs,
    initial_temp=10000.0,
    cooling_rate=0.995,
    max_iter=10000
)

end_time = time.time()
print("\nFinal Best Result:")
print(f"Best combination of nodes (simulated annealing): {best_nodes}")
print(f"Minimum cost: {min_cost}")
print(f"Time taken: {end_time - start_time:.2f} seconds")
```

```python
import networkx as nx
import numpy as np
import random

def compute_total_edge_weight(G, nodes_set):
    # Create a subgraph induced by the selected nodes
    subG = G.subgraph(nodes_set)
    # Sum the weights of all edges in the subgraph
    total_weight = sum(
        data.get('weight', 1) for _, _, data in subG.edges(data=True)
    )
    return total_weight

def cost_function(G_pi_times, G_sensitivities, nodes_set):
    sum_pi = compute_total_edge_weight(G_pi_times, nodes_set)
    sum_sen = compute_total_edge_weight(G_sensitivities, nodes_set)
    return sum_pi * sum_sen

def simulated_annealing(G_pi_times, G_sensitivities, d, max_iter=50000, initial_temp=10000, cooling_rate=0.001):
    nodes = list(G_pi_times.nodes())
    # Random initial solution
    current_solution = set(random.sample(nodes, d))
    current_cost = cost_function(G_pi_times, G_sensitivities, current_solution)
    best_solution = current_solution.copy()
    best_cost = current_cost
    temperature = initial_temp

    for iteration in range(max_iter):
        # Generate neighboring solution
        non_selected_nodes = list(set(nodes) - current_solution)
        if not non_selected_nodes:
            break
        node_to_remove = random.choice(tuple(current_solution))
        node_to_add = random.choice(non_selected_nodes)
        new_solution = current_solution.copy()
        new_solution.remove(node_to_remove)
        new_solution.add(node_to_add)
        new_cost = cost_function(G_pi_times, G_sensitivities, new_solution)

        # Acceptance probability
        delta_cost = new_cost - current_cost
        if delta_cost < 0 or random.uniform(0, 1) < np.exp(-delta_cost / temperature):
            current_solution = new_solution
            current_cost = new_cost
            # Update best solution
            if current_cost < best_cost:
                best_solution = current_solution.copy()
                best_cost = current_cost

        # Cool down
        temperature *= (1 - cooling_rate)

        # Optional: Print progress every 1000 iterations
        if iteration % 1000 == 0 and iteration > 0:
            print(f"Iteration {iteration}, Current Cost: {current_cost}, Best Cost: {best_cost}")

        # Terminate if temperature is very low
        if temperature < 1e-8:
            break

    print(f"\nBest combination of nodes found: {best_solution}")
    print(f"Minimum cost: {best_cost}")

    return best_solution, best_cost

def compute_node_sensitivities(G_sensitivities, nodes_set):
    node_sensitivities = {}
    # Create a subgraph induced by the selected nodes
    subG = G_sensitivities.subgraph(nodes_set)
    for node in nodes_set:
        # Sum the weights of all edges connected to the node within the subgraph
        total_sensitivity = sum(
            data.get('weight', 1) for _, _, data in subG.edges(node, data=True)
        )
        node_sensitivities[node] = total_sensitivity
    return node_sensitivities

# Example usage:
# Assuming you have G_pi_times and G_sensitivities already created
# Set the value of d
d = 12
best_nodes, min_cost = simulated_annealing(G_pi_times, G_sensitivities, d)

# Compute sensitivities for the selected nodes, considering only edges between them
node_sensitivities = compute_node_sensitivities(G_sensitivities, best_nodes)

ordered_nodes = sorted(node_sensitivities.items(), key=lambda item: item[1])

print("\nSelected nodes ordered by ascending sensitivity:")
for node, sensitivity in ordered_nodes:
    print(f"Node: {node}, Total Sensitivity: {sensitivity}")
```

```python

"""
A robust script to generate a qudit Ramsey pulse sequence for Ba137 iontrapping experiments.
The script accepts a list of state strings (with the first state encoded in the S manifold)
and returns a dictionary containing the initial state, U1 pulse train (with fractions, phase masks),
probe transitions, and U2 pulse train.
"""

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import ast
from collections import deque
import matplotlib.colors as mcolors

def plot_coupling_graph_from_matrix(matrix, transition_strengths, pos=None, show_graph=True):
    """
    Plot a bipartite coupling graph from the given transition matrix.
    
    Parameters:
      matrix: 2D numpy array (e.g. your pi_times matrix)
      transition_strengths: 2D numpy array of the same shape as matrix
      pos: Optional dictionary of positions for nodes.
      show_graph: If True, the graph is plotted.
      
    Returns:
      G: a networkx.Graph with nodes and weighted edges.
    """
    rows, cols = matrix.shape

    # Build row labels for the S manifold (using F=1,2,3,4 as in your example)
    Fs = [1, 2, 3, 4]
    row_labels = []
    for i in Fs:
        for j in range(2 * i + 1):
            mF = i - j
            row_labels.append([i, mF])
    # Column labels for the S states (e.g. mF values 2, -1, 0, 1, 2)
    col_labels = [-2, -1, 0, 1, 2]

    # Create graph and add edges for nonzero matrix elements.
    G = nx.Graph()
    for i in range(rows):
        for j in range(cols):
            if matrix[i, j] != 0 and not np.isnan(matrix[i, j]):
                G.add_edge(f"{row_labels[i]}", f"{col_labels[j]}", weight=transition_strengths[i, j])
    
    # Set positions (if not provided)
    if pos is None:
        pos = {
            '-2': (-2, -2),
            '-1': (-1, -2),
            '0': (0, -2),
            '1': (1, -2),
            '2': (2, -2),
        }
    # For each row label, assign a position (vertical position based on F)
    for i in row_labels:
        pos[f"{i}"] = (i[1], abs(i[0] - 4))
    # Ensure every column label is in pos
    for j in col_labels:
        if str(j) not in pos:
            pos[str(j)] = (j, -2)
        G.add_node(f"{j}", pos=pos[str(j)])
    for i in row_labels:
        G.add_node(f"{i}", pos=pos[f"{i}"])
    
    if show_graph:
        colors = list(mcolors.TABLEAU_COLORS.values())
        plt.figure(figsize=(10, 7))
        for j in range(cols):
            edges = [(f"{row_labels[i]}", f"{col_labels[j]}") 
                     for i in range(rows) if matrix[i, j] != 0]
            nx.draw_networkx_edges(G, pos, edgelist=edges,
                                   edge_color=colors[j % len(colors)], width=2, alpha=0.6)
        # Draw nodes as small rectangles
        ax = plt.gca()
        xh, yh = 0.5, 0.25
        for node, (x, y) in pos.items():
            ax.add_patch(Rectangle((x - xh / 2, y - yh / 2), xh, yh, fill=True,
                                   color='black', ec='black', lw=1))
        nx.draw_networkx_labels(G, pos, font_size=10, font_color='white')
        plt.xlim(-4.5, 4.5)
        plt.ylim(-2.5, 3.5)
        plt.title('Coupling Graph from Matrix')
        plt.show()
    
    return G

def find_least_weight_paths(start_node, end_node, matrix, weight_matrix, min_weight=0):
    """
    Using a breadth-first search on the coupling graph, find all paths from start_node to end_node
    that have the least total weight.
    
    Returns:
      A list of tuples (path, total_weight, edge_weights).
    """
    # Get the graph (no plot)
    G = plot_coupling_graph_from_matrix(matrix, weight_matrix, show_graph=False)
    queue = deque([[start_node]])
    least_weight_paths = []
    best_weight = None

    while queue:
        path = queue.popleft()
        node = path[-1]
        current_weights = []
        for i in range(len(path) - 1):
            current_weights.append(G[path[i]][path[i+1]]['weight'])
        current_total_weight = np.sum(current_weights)
        
        # Prune if this path is already heavier than the best found
        if best_weight is not None and current_total_weight > best_weight:
            continue
        
        if node == end_node:
            if best_weight is None or current_total_weight < best_weight:
                best_weight = current_total_weight
                least_weight_paths = [(path, current_total_weight, current_weights)]
            elif current_total_weight == best_weight:
                least_weight_paths.append((path, current_total_weight, current_weights))
        else:
            for neighbor in G.neighbors(node):
                if neighbor not in path:  # avoid cycles
                    edge_weight = G[node][neighbor]['weight']
                    if edge_weight >= min_weight:
                        estimated_total_weight = current_total_weight + edge_weight
                        if best_weight is None or estimated_total_weight <= best_weight:
                            new_path = list(path)
                            new_path.append(neighbor)
                            queue.append(new_path)
    return least_weight_paths

def get_transitions_from_path(path):
    """
    Given a path (a list of states), convert it into a list of transitions.
    Each transition is represented as a twoelement list: [initial_state, final_state].
    """
    transitions = []
    for i in range(len(path) - 1):
        transitions.append([path[i], path[i+1]])
    return transitions

def generate_qudit_ramsey_sequence(state_list, pi_times, transition_strengths, probe_trans_dict):
    """
    Generate the qudit Ramsey pulse sequence from the given list of states.
    
    Parameters:
      state_list: List of state strings.
          The first state should be a S state (e.g. '-2')
          and subsequent states can be either S (e.g. '-1') or D (e.g. '[4, -4]').
      pi_times: A numpy array for the pi_times matrix.
      transition_strengths: A numpy array for the transition strengths.
      probe_trans_dict: Dictionary mapping (F, mF) tuples to S state info (used for probe transitions).
      
    Returns:
      A dictionary with keys:
         'initial_state', 'pulse_train_U1', 'fractions_U1', 'simulated_phase_mask_U1',
         'fixed_phase_mask_U1', 'probe_trans', 'pulse_train_U2', 'fractions_U2',
         'simulated_phase_mask_U2', 'fixed_phase_mask_U2'
    """
    # --- Parse the input state strings ---
    parsed_states = []
    for s in state_list:
        s = s.strip()
        if s.startswith('[') and s.endswith(']'):
            try:
                parsed_states.append(ast.literal_eval(s))
            except Exception as e:
                raise ValueError(f"Error parsing state {s}: {e}")
        else:
            try:
                parsed_states.append(int(s))
            except Exception as e:
                raise ValueError(f"Error parsing state {s}: {e}")
    
    d = len(parsed_states)
    ket_0 = parsed_states[0]
    other_kets = parsed_states[1:]
    
    # --- Find transition paths from the initial S state to each target ---
    paths = []
    start_node = str(ket_0)
    for target in other_kets:
        end_node = str(target)
        candidate_paths = find_least_weight_paths(start_node, end_node, matrix=pi_times, weight_matrix=pi_times)
        if not candidate_paths:
            raise ValueError(f"No valid path found from {start_node} to {end_node}")
        # Choose the path with minimal total weight (using the second tuple element)
        best_path_tuple = min(candidate_paths, key=lambda x: x[1])
        # Convert the string labels back into the expected types:
        try:
            path_converted = [ast.literal_eval(item) if item.startswith('[') else int(item)
                              for item in best_path_tuple[0]]
        except Exception:
            path_converted = best_path_tuple[0]
        paths.append(path_converted)
    
    # --- (Optional) Sort paths; here we separate those with length > 2 from simpler ones ---
    with_third = [p for p in paths if len(p) > 2]
    without_third = [p for p in paths if len(p) <= 2]
    with_third_sorted = sorted(with_third, key=lambda x: x[2] if len(x) > 2 else 0)
    sorted_paths = without_third + with_third_sorted

    # --- Determine intermediate S transitions based on the sorted paths ---
    single_digit_states = []
    for path in sorted_paths:
        for state in path[:-1]:
            if isinstance(state, int) and state not in single_digit_states:
                single_digit_states.append(state)
    S12_paths = []
    for i in range(len(single_digit_states) - 1):
        s_node = str(single_digit_states[i])
        e_node = str(single_digit_states[i+1])
        candidate_paths = find_least_weight_paths(s_node, e_node, matrix=pi_times, weight_matrix=pi_times)
        if not candidate_paths:
            raise ValueError(f"No valid S path found from {s_node} to {e_node}")
        best_S12_path = min(candidate_paths, key=lambda x: x[1])
        try:
            converted = [ast.literal_eval(item) if item.startswith('[') else int(item)
                         for item in best_S12_path[0]]
        except Exception:
            converted = best_S12_path[0]
        S12_paths.append(converted)
    
    # --- Build the U1 pulse sequence using a heuristic to insert bus states etc. ---
    new_paths = []
    bus_inserted_once = set()
    S12_state_reached = []
    stimulated_phase_mask = []
    fixed_phase_mask = []
    fractions = []
    n = 1
    
    for path in sorted_paths:
        if len(path) > 3:
            # In this branch we assume a structure [initial_state, bus_state, transition_state, next_state]
            bus_state = tuple(path[1]) if isinstance(path[1], list) else (path[1],)
            initial_state = path[0]
            transition_state = path[2]
            # If possible, look up a matching bus from the S paths:
            matching = [sub for sub in S12_paths if len(sub) > 2 and sub[2] == transition_state]
            if matching:
                bus_state = tuple(matching[0][1]) if isinstance(matching[0][1], list) else (matching[0][1],)
                initial_state = matching[0][0]
            if bus_state not in bus_inserted_once:
                new_paths.append([initial_state, list(bus_state)])
                fractions.append(1)
                stimulated_phase_mask.append(0)
                fixed_phase_mask.append(0)
                bus_inserted_once.add(bus_state)
            if transition_state not in S12_state_reached:
                new_paths.append([transition_state, list(bus_state)])
                fractions.append(1)
                stimulated_phase_mask.append(0)
                fixed_phase_mask.append(1)
                # If the transition state is one of the targets, add an extra pulse
                if transition_state in other_kets:
                    new_paths.append([transition_state, list(bus_state)])
                    stimulated_phase_mask.append(n-1)
                    fixed_phase_mask.append(0)
                    fractions.append(1 / (d + 1 - n))
                    n += 1
                S12_state_reached.append(transition_state)
            if str(path[3]) in [str(s) for s in other_kets]:
                new_paths.append([transition_state, path[3]])
                fractions.append(1 / (d + 1 - n))
                stimulated_phase_mask.append(n-1)
                fixed_phase_mask.append(0)
                n += 1
        elif len(path) == 3:
            start_s = path[0]
            bus_s = path[1]
            end_s = path[2]
            new_paths.append([start_s, bus_s])
            stimulated_phase_mask.append(n-1)
            fixed_phase_mask.append(0)
            fractions.append(1 / (d + 1 - n))
            n += 1
            new_paths.append([end_s, bus_s])
            stimulated_phase_mask.append(0)
            fixed_phase_mask.append(1)
            fractions.append(1)
        else:
            simple_state = tuple(path[1]) if isinstance(path[1], list) else (path[1],)
            if simple_state not in bus_inserted_once:
                new_paths.append(path)
                fractions.append(1 / (d + 1 - n))
                stimulated_phase_mask.append(n-1)
                fixed_phase_mask.append(0)
                n += 1

    # --- Build the list of probe transitions using probe_trans_dict ---
    probe_trans_list = []
    for p in new_paths:
        # Here we assume that p[1] is a list [F, mF] or an int; adjust as needed.
        key = (p[1][0], p[1][1]) if isinstance(p[1], list) and len(p[1]) >= 2 else (p[1],)
        if key in probe_trans_dict:
            best_S12 = probe_trans_dict[key][0]
            if [best_S12, p[1]] not in probe_trans_list:
                probe_trans_list.append([best_S12, p[1]])

    def get_trans(path):
        """Convert a path (list of states) to a transition (list with two elements)."""
        return get_transitions_from_path(path)
    
    # --- Build the U1 pulse train from the new paths ---
    U1_pulse_train = [get_trans(p)[0] for p in new_paths if get_trans(p)]
    
    # --- Build the U2 pulse train by reversing U1 (with some adjustments) ---
    new_paths_U2 = []
    stimulated_phase_mask_U2 = []
    fixed_phase_mask_U2 = []
    fractions_U2 = []
    reversed_U1 = new_paths[::-1]
    n_U2 = d - 1
    bus_int = 0
    for i, path in enumerate(reversed_U1):
        state_repeated = False
        bus_comes_down = False
        # A simple heuristic: if the bus state appears (using the tuple representation)
        if i > 1:
            if tuple(path[1]) in bus_inserted_once and bus_int == 0:
                state_repeated = True
                bus_int = 1
            elif bus_int == 1:
                bus_comes_down = True
                bus_int = 0
        if state_repeated:
            new_paths_U2.append(path)
            stimulated_phase_mask_U2.append(0)
            fixed_phase_mask_U2.append(0)
            fractions_U2.append(1)
        elif bus_comes_down:
            new_paths_U2.append(path)
            stimulated_phase_mask_U2.append(0)
            fixed_phase_mask_U2.append(1)
            fractions_U2.append(1)
            continue
        else:
            new_paths_U2.append(path)
            stimulated_phase_mask_U2.append(n_U2)
            fixed_phase_mask_U2.append(1)
            fractions_U2.append(1 / (d + 1 - n_U2))
            n_U2 -= 1

    U2_pulse_train = [get_trans(p)[0] for p in new_paths_U2 if get_trans(p)]
    
    def flatten_transition(trans):
        """
        Flattens a transition list.
        For example, if trans is [-2, [3, -3]], it returns [-2, 3, -3].
        """
        flat = []
        for item in trans:
            if isinstance(item, list):
                flat.extend(item)
            else:
                flat.append(item)
        return flat

    # After generating your sequence dictionary, update the output transitions:
    sequence_dict = {
        'initial_state': [flatten_transition(U1_pulse_train[0])] if U1_pulse_train else [],
        'pulse_train_U1': [flatten_transition(item) for item in U1_pulse_train],
        'fractions_U1': fractions,
        'simulated_phase_mask_U1': [0]*len(fractions),
        'fixed_phase_mask_U1': fixed_phase_mask,
        'probe_trans': [flatten_transition(item) for item in probe_trans_list],
        'pulse_train_U2': [flatten_transition(item) for item in U2_pulse_train],
        'fractions_U2': fractions_U2,
        'simulated_phase_mask_U2': stimulated_phase_mask_U2,
        'fixed_phase_mask_U2': fixed_phase_mask_U2
    }

    
    return sequence_dict


# Example input states for a D=3 Ramsey (first state is S, others are D)
input_states = ['-2', '[3, -3]', '[3, -1]', '-1', '[2, -2]', '[4, -4]', '[2, -1]']

sequence = generate_qudit_ramsey_sequence(input_states, pi_times,
                                            transition_strengths, probe_trans_dict)
print("Generated Qudit Ramsey Sequence:")
for key, value in sequence.items():
    print(f"{key} = {value}")
```

```python
from PIL import Image
import io
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from IPython import display
import time
import numpy as np

def animate_paths_with_population_transfer_bidirectional(list_of_paths, transfer_fractions, transfer_phase, initial_population, 
                                                          delay=0.5, save_path='animation.gif'):

    # Store frames for GIF
    frames = []
    
    G = nx.DiGraph()
    print(list_of_paths, transfer_fractions, transfer_phase)
    # Setup node positions
    Fs=[0, 1, 2, 3, 4]
    row_labels = []
    for i in Fs:
        for j in range(2 * i + 1):
            mF = i - j
            row_labels.append([i, mF])
    
    pos = {
        '-2': (-2, -2),
        '-1': (-1, -2),
        '0': (0, -2),
        '1': (1, -2),
        '2': (2, -2),
    }
    for i in row_labels:
        pos[f"{i}"] = (i[1], abs(i[0] - 4))

    # Add all nodes with positions
    for node in pos.keys():
        G.add_node(node, pos=pos[node])
    
    # Convert paths to strings
    string_paths = [[str(node) if not isinstance(node, list) else str(node) for node in path] 
                    for path in list_of_paths]
    print('string_paths = ',string_paths)
    # Initialize population tracking
    node_populations = {}  # Track current population of each node
    population_history = {}  # Track population changes over time
    node_amplitudes = {}  # Complex amplitudes
    # Set initial population
    start_node = string_paths[0][0]  # Assume all paths start from same node
    print(start_node)
    node_amplitudes[start_node] = np.sqrt(initial_population)

    # Create list of all population transfer events
    transfer_events = []
    print(transfer_fractions, transfer_phase)
    for path, fraction, phase in zip(string_paths, transfer_fractions, transfer_phase):
        from_node = path[0]
        to_node = path[1]
        if to_node not in node_amplitudes:
            node_amplitudes[to_node] = 0
        print(path, fraction, phase)
        theta = np.arcsin(np.sqrt(fraction)) 
        phi = phase

        # Givens rotation matrix
        rotation_matrix = np.array([
            [np.cos(theta), -1j*np.sin(theta) * np.exp(-1j * np.pi*phi)],
            [-1j*np.sin(theta) * np.exp(1j * np.pi*phi), np.cos(theta)]
        ])
        # hamiltonian = np.array([
        #     [0,theta*np.exp(-1j*np.pi*phi)],
        #     [theta*np.exp(1j*np.pi*phi),0]
        # ])
        # rotation_matrix = np.exp(-hamiltonian)
        # Apply rotation
        from_amp = node_amplitudes.get(from_node, 0.0)
        to_amp = node_amplitudes.get(to_node, 0.0)

        updated_amplitudes = rotation_matrix @ np.array([from_amp, to_amp])
        node_amplitudes[from_node] = updated_amplitudes[0]
        node_amplitudes[to_node] = updated_amplitudes[1]

        # Store the event
        transfer_events.append({
            'edge': (from_node, to_node),
            'from_amp': node_amplitudes[from_node],
            'to_amp': node_amplitudes[to_node],
            'theta': theta,
            'phi': phi,
        })

        # Store in history
        
        for node, amp in node_amplitudes.items():
            if node not in population_history:
                population_history[node] = {}
            step = len(transfer_events) - 1
            population_history[node][step] = np.round(abs(amp)**2,2)  # Store population (amplitude squared)

     
    def capture_frame():
        # Save current figure to a buffer
        buf = io.BytesIO()
        plt.savefig(buf, format='png', bbox_inches='tight', dpi=300)
        buf.seek(0)
        return Image.open(buf)

    def draw_frame(step, current_edge=None, node_colors=None):
        plt.figure()  
        plt.clf()
        ax = plt.gca()
        
        # Set default node colors if not provided
        if node_colors is None:
            node_colors = {node: 'black' for node in pos.keys()}
        
        # Draw nodes with the specified colors
        xh = 0.5
        yh = 0.25
        for node, (x, y) in pos.items():
            # Use the specified color for each node
            color = node_colors.get(node, 'black')
            ax.add_patch(Rectangle((x - xh / 2, y - yh / 2), xh, yh, 
                                   fill=True, color=color, ec='black', lw=1))
            
            # Add population text below node if it has a population
            if node in population_history and step in population_history[node]:
                plt.text(x, y - yh/2 - 0.1, f'({population_history[node][step]})', 
                         ha='center', va='top', color='black',fontsize = 8)
        
        # Draw single current edge if exists
        if current_edge:
            temp_graph = nx.DiGraph()
            temp_graph.add_nodes_from(G.nodes(data=True))
            temp_graph.add_edge(*current_edge)
            nx.draw_networkx_edges(
                temp_graph, pos, alpha=1, edge_color='red', 
                arrowstyle='<|-|>', arrowsize=15
            )
        
        # Draw node labels
        nx.draw_networkx_labels(G, pos, font_size=7, font_color='white')
        
        plt.xlim(-4.5, 4.5)
        plt.ylim(-2.5, 3.5)
        plt.title('Transition Paths Animation with Bidirectional Population Transfer',fontsize = 8)
        
        # Capture frame for GIF
        frames.append(capture_frame())

    # Initial frame (no edges)
    draw_frame(0)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(delay)
    print(len(transfer_events))
    # Animate edges and population changes one at a time
    for i, event in enumerate(transfer_events):
        # Highlight nodes involved in the current transfer
        from_node, to_node = event['edge']
        node_colors = {node: 'black' for node in pos.keys()}  # Default all nodes to black
        node_colors[from_node] = 'blue'  # Highlight source node
        node_colors[to_node] = 'blue'  # Highlight destination node
        
        # Draw frame with highlighted nodes
        draw_frame(i, current_edge=event['edge'], node_colors=node_colors)
        display.clear_output(wait=True)
        display.display(plt.gcf())
        time.sleep(delay)
        
        # Show frame without edge but with updated populations
        if i < len(transfer_events) - 1:  # Don't need extra frame after last transfer
            draw_frame(i)
            display.clear_output(wait=True)
            display.display(plt.gcf())
            time.sleep(delay)
    
    # Save the animation as GIF
    frames[0].save(
        save_path,
        save_all=True,
        append_images=frames[1:],
        duration=int(delay * 1000),  # PIL uses milliseconds
        loop=0
    )
    
    # plt.show()
```
